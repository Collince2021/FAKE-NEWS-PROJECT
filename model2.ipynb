```python
import pandas as pd
import numpy as np
import re
import string
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df1=pd.read_csv("True Clean.csv")
df2=pd.read_csv("Fake Clean.csv")
```

```python
##getting reuired columns
Real_Newdata1=df1[["text"]]
Fake_Newdata1=df2[["text"]]
```

```python
##adding classification column
Real_Newdata1["class"]=1
Fake_Newdata1["class"]=0
```

```python
Real_Newdata1.head(3)
```

```python
##merge real and fake datasets
df=pd.concat([Real_Newdata1,Fake_Newdata1],axis=0)
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
##reshuffle the dataset
df=df.sample(frac=1)
```

```python
df.head(10)
```

```python
import nltk
from nltk.tokenize import word_tokenize ##for getting tokens out of words
from nltk.stem import WordNetLemmatizer  ##for converting the words to their root form
nltk.download('punkt')
nltk.download('wordnet')


def preprocess(text):
    ##replacing whitespace characters with a single space
   ## text=re.sub("\s"," ",text)
    ##perfoming lemmatization
    
    # Initialize the WordNetLemmatizer
    lemmatizer = WordNetLemmatizer()
    # Tokenize the text into words
    tokens = word_tokenize(text)
    # Lemmatize each word in the text
    lemmatized_text = [lemmatizer.lemmatize(word) for word in tokens]
    # Join the lemmatized words back into a single string
    lemmatized_text = ' '.join(lemmatized_text)
    return lemmatized_text

```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer

# Split the data into features text and labels
X = df['text']
y = df['class']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Define the pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),  # Text vectorization(perfoms text cleaning)
    ('clf', RandomForestClassifier())  # Random Forest classifier
])

# Fit the pipeline to the training data
pipeline.fit(X_train, y_train)

# Predict labels for the testing data
y_pred = pipeline.predict(X_test)
```

```python
##evaluatng the model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(confusion_matrix(y_test,y_pred))
```
